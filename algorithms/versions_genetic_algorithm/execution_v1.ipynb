```python
from visualization import plot_map, plot_stats

import genetic_v1 as genetic 
# Subject, Genetic, Mapa, Gene, GeneDecoded, CartesianPoint
```

```python
wp_ori = genetic.CartesianPoint(10.0, 10.0)
wp_des = genetic.CartesianPoint(40.0, 10.0)

verts1 = [
    (20.0, 5.0), # left, bottom
    (20.0, 15.5), # left, top
    (30.0, 15.5), # right, top
    (30.0, 5.0), # right, bottom
    (20.0, 5.0), # ignored
]
verts1 = [genetic.CartesianPoint(v[0], v[1]) for v in verts1]


mapa1 = genetic.Mapa(wp_ori, wp_des, [verts1])

ag = genetic.Genetic(genetic.Subject, mapa1,
        C_d=100,
        C_obs=1000,
        C_con=1,
        C_cur=1
)

best = ag.run(max_exec_time=40, info=True)
print('-'*30)
print(best.dna)
print('-'*30)
print(best.dna_decoded)
best = ag.best

plot_map(
    [verts1], 
    ['n'], 
    wp_ori, 
    wp_des, 
    best.get_route(),
    None
)
plot_stats(ag.trace, normalize=True)
plot_stats(ag.trace, normalize=False)
```
